### downloading the files from URL

```python
import pandas as pd
import requests
import os

# Load the CSV file containing document URLs
csv_file = 'IR_corpus_links.csv'
df = pd.read_csv(csv_file)

# Directory to save downloaded documents
download_dir = 'downloaded_documents'
os.makedirs(download_dir, exist_ok=True)

# Function to download documents
def download_documents(df, download_dir):
    for index, row in df.iterrows():
        doc_url = row['Link']
        
        # Skip rows with NaN or invalid URLs
        if pd.isna(doc_url) or not isinstance(doc_url, str):
            print(f"Skipping row {index} due to invalid URL.")
            continue
        
        doc_name = os.path.join(download_dir, doc_url.split('/')[-1])
        print(f"Downloading {doc_name} from {doc_url}...")
        
        try:
            with requests.get(doc_url, stream=True) as r:
                r.raise_for_status()
                with open(doc_name, 'wb') as f:
                    for chunk in r.iter_content(chunk_size=8192):
                        f.write(chunk)
            print(f"Downloaded {doc_name}")
        except Exception as e:
            print(f"Failed to download {doc_name}: {e}")

# Download documents
download_documents(df, download_dir)
```

### extract text from each document and convert to markdown

```python
import os
import PyPDF2

# Directories
input_dir = 'downloaded_documents'  # Replace with the path to your PDF directory
text_output_dir = 'pdf2text'  # Replace with the path to your text output directory
markdown_output_dir = 'text2md_files'  # Replace with the path to your markdown output directory

# Ensure output directories exist
os.makedirs(text_output_dir, exist_ok=True)
os.makedirs(markdown_output_dir, exist_ok=True)

def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF file."""
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ''
            for page in reader.pages:
                text += page.extract_text()
            return text
    except Exception as e:
        print(f"Error processing {pdf_path}: {e}")
        return None

def save_text_and_md(pdf_file, input_dir, text_output_dir, markdown_output_dir):
    """Extracts text from a PDF and saves as both .txt and .md files."""
    try:
        pdf_path = os.path.join(input_dir, pdf_file)
        text = extract_text_from_pdf(pdf_path)
        if text:
            base_filename = os.path.splitext(pdf_file)[0]
            
            # Save as .txt
            txt_filepath = os.path.join(text_output_dir, base_filename + '.txt')
            with open(txt_filepath, 'w', encoding='utf-8') as txt_file:
                txt_file.write(text)
            print(f"Saved text file: {txt_filepath}")
            
            # Save as .md
            md_filepath = os.path.join(markdown_output_dir, base_filename + '.md')
            with open(md_filepath, 'w', encoding='utf-8') as md_file:
                md_file.write(text)
            print(f"Saved markdown file: {md_filepath}")
    except Exception as e:
        print(f"Error saving files for {pdf_file}: {e}")

# Process all PDF files
for pdf_file in os.listdir(input_dir):
    if pdf_file.lower().endswith('.pdf'):
        save_text_and_md(pdf_file, input_dir, text_output_dir, markdown_output_dir)
```

```python
import os
import pathlib
import pymupdf4llm

# Directories
input_dir = 'downloaded_documents'  # Replace with the path to your PDF directory
markdown_output_dir = 'pdf2md_files'  # Replace with the path to your markdown output directory

# Ensure output directory exists
os.makedirs(markdown_output_dir, exist_ok=True)

def convert_pdf_to_md(pdf_file, input_dir, markdown_output_dir):
    """Converts a PDF to Markdown and saves it as a .md file."""
    try:
        pdf_path = os.path.join(input_dir, pdf_file)
        
        # Use pymupdf4llm to convert PDF to Markdown text
        md_text = pymupdf4llm.to_markdown(pdf_path)
        
        # Create the output markdown file path
        base_filename = os.path.splitext(pdf_file)[0]
        md_filepath = os.path.join(markdown_output_dir, base_filename + '.md')
        
        # Save the Markdown text to a .md file
        pathlib.Path(md_filepath).write_bytes(md_text.encode('utf-8'))
        print(f"Saved Markdown file: {md_filepath}")
    except Exception as e:
        print(f"Error converting {pdf_file} to Markdown: {e}")

# Process all PDF files
for pdf_file in os.listdir(input_dir):
    if pdf_file.lower().endswith('.pdf'):
        convert_pdf_to_md(pdf_file, input_dir, markdown_output_dir)
```
